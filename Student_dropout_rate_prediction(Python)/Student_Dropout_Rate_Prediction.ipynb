# Student Dropout Rate Prediction Project

### Problem Statement

Educational institutions face challenges in identifying students at risk of dropping out. Early detection can help institutions provide timely interventions and improve academic success rates.

This project aims to perform Exploratory Data Analysis (EDA) and build a machine learning classification model to predict whether a student will:

- Drop out

- Likely to Graduate

### Goal

- Target

The Target variable indicates the final academic outcome of a student. The goal of this project is to develop a supervised machine learning model capable of accurately classifying students into one of the three categories: Dropout, Enrolled, or Graduate, using the available input features.

### EDA

```python
# Display skewness of numerical features
numerical_cols = ['Previous qualification (grade)',
                  'Admission grade', 'Age at enrollment',
                  'Curricular units 1st sem (grade)',
                  'Curricular units 2nd sem (grade)',
                  'Unemployment rate', 'Inflation rate', 'GDP']

for col in numerical_cols:
    skewness = df[col].skew()
    print(f'Skewness of {col}: {skewness}')
```

```python
# Percentage of Dropouts
dropout_percentage = df['Target'].value_counts(normalize=True).iloc[1] * 100
print(f"Percentage of Dropouts: {dropout_percentage:.2f}%")
```

```python
# percentage of students who dropped out grouped by different categories
cols = ['Marital status', 'Application mode', 'Course',
       'Daytime/evening attendance', 'Previous qualification',
       'Nationality','Displaced',
       'Educational special needs', 'Debtor',
       'Gender', 'Scholarship holder', 'International']
for col in cols:
    dropout_by_category = df.groupby(col)['Target']\
        .value_counts(normalize=True)\
        .unstack() * 100
    dropout_by_category.fillna(0, inplace=True)
    print(dropout_by_category.iloc[:, 0].map('{:.2f}%'.format))
    print("-------------------------------")
```

### Preprocessing

```python
df.nunique()
```

```python
# One-hot encoding of Low Cardinality categorical columns
low_cardinality_cols = ['Marital status', 'Application order']

ohe = OneHotEncoder(dtype=int)

encoded = ohe.fit_transform(df[low_cardinality_cols])

encoded_df = pd.DataFrame.sparse.from_spmatrix(
    encoded,
    columns=ohe.get_feature_names_out(low_cardinality_cols),
    index=df.index
)

df = pd.concat([df.drop(columns=low_cardinality_cols), encoded_df], axis=1)

df.head()
```

### Model Building

```python
# Train-test split
x = df.drop('Target', axis=1)
y = df['Target']
x_train, x_test, y_train, y_test = train_test_split\
    (x, y, test_size=0.2, random_state=42)
```

#### Decision Tree

```python
# Training the Decision Tree Classifier
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(x_train, y_train)
```

```python
# Predicting the target variable for the test set
y_pred_dtc = dtc.predict(x_test)
```

```python
# Checking the accuracy of the Decision Tree Classifier
accuracy_dtc = accuracy_score(y_test, y_pred_dtc)
confusion_dtc = confusion_matrix(y_test, y_pred_dtc)
print(f'Accuracy of Decision Tree Classifier: {(accuracy_dtc * 100):.2f}%')
print(f'Confusion Matrix of Decision Tree Classifier:\n{confusion_dtc}')
```

#### Random Forest 

```python
# Training the Random Forest Classifier
rfc=RandomForestClassifier(random_state=42)
rfc.fit(x_train, y_train)
```

```python
# Predicting the target variable for the test set
y_pred_rfc = rfc.predict(x_test)
```

```python
# Checking the accuracy of the Random forest Classifier
accuracy_rfc = accuracy_score(y_test, y_pred_rfc)
confusion_rfc = confusion_matrix(y_test, y_pred_rfc)
print(f'Accuracy of Random Forest Classifier: {(accuracy_rfc * 100):.2f}%')
print(f'Confusion Matrix of Random Forest Classifier:\n{confusion_rfc}')
```
